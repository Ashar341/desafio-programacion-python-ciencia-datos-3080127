#primero quitamos la columna appellidos, otra manera seria usar columns = 'Apellido'

```python
df2 = df.drop('Apellido', axis=1)
df2
```

Ahora se Dejaran las lineas que tengan mas de 5 valores validos

```python
df4 = df2.dropna(axis=0, thresh=5)
df4.shape
df4
```

De esta manera encontramos que tenemos 245  lineas con datos con mas de 5 columnas con valores y 145 completos del total de 251, ahora faltaria revisar si no hay valores duplicados y eliminarlos del registro bueno, que en este caso es el df3

```python
print(df3.shape)
duplicados = df3.duplicated()
print(df3[duplicados]) # de esta manera sale en tabla

df5 = df3.drop_duplicates()
df5
```
