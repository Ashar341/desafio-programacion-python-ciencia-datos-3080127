```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage 
from sklearn.preprocessing import normalize, StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
import scipy.cluster.hierarchy as shc
from sklearn.metrics import silhouette_score
```

Paso 2 Agregar y limpiar los datos

```python
df = pd.read_csv('clientes.csv', delimiter=";")
```

```python
print(df.head())
```

Paso 3 Procesar los datos

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
 
# Normalizing the data so that the data approximately
# follows a Gaussian distribution
X_normalized = normalize(X_scaled)
 
# Converting the numpy array into a pandas DataFrame
X_normalized = pd.DataFrame(X_normalized)
X_normalized
```

Paso 4 reducir la dimensionalidad

```python
pca = PCA(n_components = 2)
X_principal = pca.fit_transform(X_normalized)
X_principal = pd.DataFrame(X_principal)
X_principal.columns = ['P1', 'P2']
X_principal
```

Paso 5 Graficar

```python
plt.figure(figsize =(8, 8))
plt.title('Visualising the data')
Dendrogram = shc.dendrogram((shc.linkage(X_principal, method ='ward')))
```

Aqui se determinan cuantos clusteres son necesarios a traves de la visualizacion, por lo que se escogeran 3
Paso 6 Vializando con diferentes clusteres


```python
ac2 = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage='ward')
ac2.fit(df)
labels = ac2.labels_

# Visualizing the clustering
plt.scatter(df['Salario'], df['Gasto'], c=labels) 
plt.xlabel('Salario Mensual')
plt.ylabel('Porcentaje Gasto Mensual')
```

```python
ac3 = AgglomerativeClustering(n_clusters = 3, affinity='euclidean', linkage='ward')
ac3.fit(df)
labels = ac3.labels_

# Visualizing the clustering
plt.scatter(df['Salario'], df['Gasto'], c=labels) 
plt.xlabel('Salario Mensual')
plt.ylabel('Porcentaje Gasto Mensual')
```

```python
ac4 = AgglomerativeClustering(n_clusters = 4, affinity='euclidean', linkage='ward')
ac4.fit(df)
labels = ac4.labels_

# Visualizing the clustering
plt.scatter(df['Salario'], df['Gasto'], c=labels) 
plt.xlabel('Salario Mensual')
plt.ylabel('Porcentaje Gasto Mensual')
```

```python
ac5 = AgglomerativeClustering(n_clusters = 5, affinity='euclidean', linkage='ward')
ac5.fit(df)
labels = ac5.labels_

# Visualizing the clustering
plt.scatter(df['Salario'], df['Gasto'], c=labels) 
plt.xlabel('Salario Mensual')
plt.ylabel('Porcentaje Gasto Mensual')
```

```python
ac6 = AgglomerativeClustering(n_clusters = 6, affinity='euclidean', linkage='ward')
ac6.fit(df)
labels = ac6.labels_

# Visualizing the clustering
plt.scatter(df['Salario'], df['Gasto'], c=labels) 
plt.xlabel('Salario Mensual')
plt.ylabel('Porcentaje Gasto Mensual')
```

Paso 7 Evaluando el cluster

```python

k = [2, 3, 4, 5, 6]
 
# Appending the silhouette scores of the different models to the list
silhouette_scores = []
silhouette_scores.append(
        silhouette_score(X_principal, ac2.fit_predict(X_principal)))
silhouette_scores.append(
        silhouette_score(X_principal, ac3.fit_predict(X_principal)))
silhouette_scores.append(
        silhouette_score(X_principal, ac4.fit_predict(X_principal)))
silhouette_scores.append(
        silhouette_score(X_principal, ac5.fit_predict(X_principal)))
silhouette_scores.append(
        silhouette_score(X_principal, ac6.fit_predict(X_principal)))
 
# Plotting a bar graph to compare the results
plt.bar(k, silhouette_scores)
plt.xlabel('Number of clusters', fontsize = 20)
plt.ylabel('S(i)', fontsize = 20)
plt.show()
```

Con lo anterior obtenemos que un cluster de 3 es lo ideal para representar nuestros datos
