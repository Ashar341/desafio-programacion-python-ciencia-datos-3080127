```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('clusters.csv', delimiter=";")
```

```python
print(df.head(10))
```

```python
plt.plot(df['Latitud'], df['Longitud'])
plt.ylabel('some numbers')
plt.show()
```

```python
kmeans = KMeans(n_clusters=10, random_state=0)
kmeans.fit(df)
y_kmeans = kmeans.predict(df)
cluster_centers = kmeans.cluster_centers_
```

```python
df['cluster'] = kmeans.predict(df)
df.sample(5, random_state = 0)
```

```python
plt.scatter(df['Latitud'], df['Longitud'], c=y_kmeans, s=20)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200)
plt.xlabel('Latitud')
plt.ylabel('Longitud') 
```

ahora se puede calcular que tan exacto es el modelo a traves del analisis de los datos de las especies, a traves de un mapeo
